```python
import requests
from bs4 import BeautifulSoup
import lxml
import re
```

```python
urls = [f'https://book24.ru/rating/year/page-{i}/' for i in range(1, 8)]

for url in urls:
    print(url)
```

```python
headers = {
    'Accept': '*/*',
    'User-Agent': ''
}
```

```python
responses = [requests.get(url).text for url in urls]
print(responses[5])
```

```python
page = responses[0]

soup = BeautifulSoup(page, 'lxml')

cards = [card.text for card in soup.find_all('a', {'class': 'product-card__name'})]

print(cards)
```

```python
card_auth = [card.text for card in soup.find_all('a', {'class': 'author-list__item smartLink'})]

print(card_auth)
```
